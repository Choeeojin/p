# 과제 제출

## 과제 git에 제출하기

1. git에 제출한다는 의미
    - 자신이 작업하고 있는 컴퓨터의 과제파일을 인터넷 git서버에 올린다는 의미
    - 자신의 컴퓨터를 **로컬**이라고 한다.

2. 준비
    - github.com 회원 가입
    - github.com에 sign in
    - dowload 'Github Desktop'
    - 다운로드 받고, 자신의 **로컬** 컴퓨터에 'Github Desktop' 설치.
    - 가장 쉬운 방법은:
        - 아래 3번 부터 생략하고, **git서버**에서 git repo를 만들고,
        - upload files메뉴를 이용해, 마우스로 과제파일을 끌어다 놓고,
        - 맨 아래 commit을 누르면 된다.

3. **로컬**에 git 폴더 (git repo) 만들기
    - 내려받은 'Github Desktop'을 실행한다.
    - 좌측 상단 +를 누르면 Add Create Clone 세가지 메뉴가 나온다.
    - 그 중에 Create를 누른다.
    - 새 폴더 이름을 입력한다.
        - 과목 1글자 + 분반 1글자 + 학번
        - 1분반 예
        ```
        p1_2016xxxxx
        ```
        
        - 2분반 예
        ```
        p2_2016xxxxx 
        ```
        
    - 그러면 **로컬에** 폴더가 생성된다 (git repo라고 하고, 그 안에 .git폴더가 있다.)

4. **로컬에** 숙제파일을 추가한다.
    - 파일을 생성하면 그 폴더 (git repo)에 가져다 놓는다.
    - 이미 있는 파일을 수정하는 경우, 그 파일을 열어서 수정한다.
    - 주차별 이름을 붙이고 뒤에 .py라고 한다.
    - 3주차 과제 명명의 예.
    
    ```
    w3Main.py
    ```
    
    - 명령창에서 w3Main.py가 저장되어 있는 폴더로 가서, 실행하면 오류가 없어야 한다.
    
    ```
    python w3Main.py
    ```

5. Github Desktop 상단 가운데 Changes History 메뉴에서 Changes를 누른다.
    - 변경된 파일이 목록에 뜬다.
    - 제출할 숙제가 목록에 있는지 확인한다.

6. **로컬**에서 숙제를 제출하기 위한 commit
    - Summary와 Description을 입력한다.
    - 그러면 하단에 Commit to master가 활성화 된다.

7. Github Desktop 우 상단 Sync를 누른다.
    - 6번에서 commit한 파일이 내 계정으로 제출된다.

8. 잘 제출되었는지 **git 서버**에 가서 확인
    - 회원가입했던 **github.com**에 가서, 자신의 계정에 그 파일이 목록에 있는지 확인한다.

9. 다음 부터 숙제를 제출할 때는 4번부터 8번까지 실행한다.

## 공지

- 폴더명, 파일명은 영어를 사용한다.
- 수업시간에 제출해야 하는 과제는 수업시간에 제출한다.
- 그 외는 강의계획서에 공지한 바와 같이 제출한다.
- git로 제출했으면, 별도로 출력물을 제출하지 않는다.
- git에는 .py만 제출하도록 한다. 이미지, 바이너리 파일은 올리지 않는다.
- w3Main.py의 예는 src/w3Main.py참조.

## 채점기준

- 문제에 적합한 정상실행 100%
- 오작동 미제출 0점 처리.
- 제출시점 미준수: 24시간 내 제출 50% 감점. 24시간 이후 0점 처리.
- 모든 과제는 반드시 제출. 미제출 비율은 성적에 반영 (20% 미제출할 경우, 성적을 1계단 내린다 예: A+ -> A)
